```python
import os
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
import pandas as pd
df = pd.read_csv('input_dataset.csv')
```

```python
df = df.drop('MU', axis=1)
```

```python
df = df.drop('categories ', axis=1)
```

```python
import os
from ragas import evaluate
import pandas as pd
from datasets import Dataset
from ragas.metrics import answer_correctness
import nest_asyncio

nest_asyncio.apply()
```

```python
evaluation_dataset = Dataset.from_pandas(df)
evaluation_dataset
```

```python
# Ensure that answer_correctness is in a list
metrics = [answer_correctness]

# Evaluate the dataset using the specified metrics
score = evaluate(evaluation_dataset, metrics=metrics)

# Convert the score to pandas DataFrame for better visualization
results = score.to_pandas()
```

```python
results.to_csv('results.csv', index=False)
results
```
